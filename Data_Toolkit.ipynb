15. How does NumPy handle multidimensional arrays?

NumPy handles multidimensional arrays using its powerful ndarray object, which supports arrays of any number of dimensions—from simple 1D vectors to complex 4D tensors and beyond. Here's how it works:

Core Concepts

- ndarray Structure: NumPy arrays are stored in contiguous memory blocks and can have shapes like (3,), (3, 4), (2, 3, 4), etc., representing 1D, 2D, 3D arrays respectively.

- Efficient Indexing: You can access elements using multi-level indexing like arr[1, 2] for 2D or arr[0, 1, 2] for 3D arrays.

- Reshaping: Use .reshape() to change the shape of an array without altering its data.

- Broadcasting: Smaller arrays are automatically expanded to match the shape of larger ones during operations.

- Mathematical Operations: Supports element-wise operations, matrix multiplication, and statistical functions across dimensions.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

NumPy’s array slicing is faster than Python’s list slicing because of how data is stored and accessed under the hood. Here's the breakdown:

- Contiguous Memory Layout NumPy arrays store data in a single, tightly packed block of memory. This allows for faster access and manipulation because the CPU can efficiently cache and retrieve data.

- Homogeneous Data Types All elements in a NumPy array share the same data type and size. This uniformity enables optimized operations and avoids the overhead of type checking that Python lists require.

- Pointer Chasing in Lists Python lists are arrays of pointers to objects, which may be scattered across memory. Accessing each element involves following a pointer, which slows down performance due to poor cache locality.

- C-Level Implementation NumPy slicing operations are implemented in highly optimized C code. This means slicing doesn’t involve Python loops or object overhead—it’s handled at a lower level with minimal latency.

- View-Based Slicing NumPy slicing often returns a view of the original array, not a copy. This avoids memory duplication and speeds up operations. Python lists, on the other hand, create new lists when sliced.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
df = sns.load_dataset("penguins")

# Step 1: Calculate the correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Step 2: Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)

# Step 3: Display the plot
plt.title("Correlation Matrix Heatmap")
plt.show()
````

11. Perform matrix multiplication using NumPy.

````python
import numpy as np

# Define two compatible matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Method 1: Using np.dot()
result_dot = np.dot(A, B)

# Method 2: Using np.matmul()
result_matmul = np.matmul(A, B)

# Method 3: Using @ operator (Python 3.5+)
result_at = A @ B

# Display results
print("Result using np.dot():\n", result_dot)
print("Result using np.matmul():\n", result_matmul)
print("Result using @ operator:\n", result_at)
````
